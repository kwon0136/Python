```python
crime_seoul = pd.read_csv("../Data/crime_Seoul2.csv", encoding='utf-8', thousands=",", index_col=0)
crime_seoul.head()
```

```python
crime_seoul['관서명']

stations=[]
for name in crime_seoul['관서명']:
    stations.append('서울' + name[:-1]+ '경찰서')

stations
```

```python
import folium
#!pip install folium
```

```python
# 위도, 경도로 위치 찾기
fm = folium.Map(location=[37.5636465, 126.9895796])
folium.Marker([37.5636465, 126.9895796], popup='중부경찰서', icon=folium.Icon(icon='cloud')).add_to(fm)
fm
```

```python
## 전체 서울경찰서를 folium.Marker를 이용하여 지도에 출력
data=[]
for station in stations:
    #print(station)
    data.append(gmaps.geocode(station, language='ko'))
    #print('')
#data[0][0].get('geometry')['location']


fm2 = folium.Map(location=[37,126], zoom_start=7) # 초기 위치 설정 필요
for i in range(len(data)):
    folium.Marker([data[i][0]['geometry']['location']['lat'],data[i][0]['geometry']['location']['lng']],popup=stations[i], icon=folium.Icon(icon='cloud')).add_to(fm2) 
fm2
```

```python
lat=[]
lng=[]
for i in range(len(data)):
    lat.append(data[i][0]['geometry']['location']['lat'])
    lng.append(data[i][0]['geometry']['location']['lng'])

m = folium.Map(location=[37,126], zoom_start=7)

for i in range(len(data)):
    folium.Marker([lat[i],lng[i]], popup=stations[i], icon=folium.Icon(icon='cloud')).add_to(m)
m
```
