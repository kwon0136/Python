```python
crime_seoul = pd.read_csv("Data/crime_Seoul2.csv", encoding='utf-8',  
                         index_col=0)

station_addresses = []
station_name = []
station_lat = []
station_lng = []
#print(crime_seoul['관서명'])
for name in crime_seoul['관서명']:
    nm = '서울' + name[:-1] + '경찰서'
    #print(nm)
    tmp = gmaps.geocode(nm, language='ko')
    address = tmp[0].get("formatted_address")    
    tmp_loc = tmp[0].get("geometry")
    lat = tmp_loc['location']['lat']
    lng = tmp_loc['location']['lng']
    station_addresses.append(address)
    station_name.append(name)
    station_lat.append(lat)
    station_lng.append(lng)
    #print(nm, ',', address, ',', lat, lng, ',', name)
print(len(station_addresses), 
      len(station_name),
      len(station_lat),
      len(station_lng))
```

```python
df = pd.DataFrame()
df['address'] = station_addresses
df['name'] = station_name
df['lat'] = station_lat
df['lng'] = station_lng
```

```python
import folium
fm = folium.Map(location=[df['lat'].mean(), df['lng'].mean()], zoom_start=11)

for i in df.index:
    folium.Marker(location=[df['lat'][i], df['lng'][i]], popup=df['name'][i]).add_to(fm)

fm
```
