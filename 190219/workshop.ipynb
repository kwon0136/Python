```python
#송도1동 이라는 이름이 포함된 지역의 인구 그래프

import csv
f = open("village_population.csv", encoding= "utf-8")
data = csv.reader(f)

po = []
for row in data:
    if '송도1동' in row[0]:
        for i in range(3,102):
            if row[i] =="":continue
            po.append(int(row[i]))
#print(po)
import matplotlib.pyplot as plt
plt.plot(po)
plt.show()
```

```python
import csv
f = open("village_population.csv", encoding="utf-8")
data = csv.reader(f)
next(data)

result = []
for row in data:
    if '송도1동' in row[0]:
        for i in range(3, len(row)):
            result.append(int(row[i]))
            
import matplotlib.pyplot as plt
plt.plot(result)
plt.show()
plt.hist(result)
plt.show()
```
