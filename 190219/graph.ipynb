```python
import csv
f = open('seoul.csv', encoding = 'utf-8')
data = csv.reader(f)
next(data)

temperature = []
for row in data:
    if(row[-1] != '' and (row[0].split("-")[1] == "11" and row[0].split("-")[2] == "24")):
        temperature.append(float(row[-1]))
        
#print(temperature)

import matplotlib.pyplot as plt
plt.plot(temperature)
plt.show()
```
