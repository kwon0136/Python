```python
import csv
f = open("seoul.csv", encoding="utf-8")
data = csv.reader(f)
next(data)

temperature=[]
for row in data:
    if (row[-1] !='' and(row[0].split("-")[1] == '11' and row[0].split('-')[2] == '24')):
        temperature.append(float(row[-1]))
        
import matplotlib.pyplot as plt
plt.plot(temperature)
plt.show()
```

```python
#역대 서울 기온 중 37 이상이었던 날의 날짜와 최고 기온 출력
import csv
f = open("seoul.csv", encoding= "utf-8")
data = csv.reader(f)
next(data)

max_temp = []
max_temp_date = []
for row in data:
    if row[-1] != '' and float(row[-1]) >= 37:
        max_temp.append(float(row[-1]))
        max_temp_date.append(row[0])
#print(max_temp_date, max_temp)
```

```python
#역대 서울의 최고 기온을 히스토그램으로 출력
import csv
f = open("seoul.csv", encoding='utf-8')
data = csv.reader(f)
next(data)

max = []
for row in data:
    if(row[-1] != ''):
        max.append(float(row[-1]))
        
import matplotlib.pyplot as plt
plt.hist(max, bins = 100)
plt.show()
```

```python
#송도1동 연령별 인구수
import csv
f = open('village_population.csv', encoding='utf-8')
data = csv.reader(f)
next(data)
data2 = list(data)  #송도1동으로 for문 수행 후 가장 마지막 데이터 값으로 가기 때문에 신도림으로 for문을 수행해도 결과값이 나오지 않는다

result = []
for row in data2:
    if "송도1동" in row[0]:
        for i in range(3, len(row)):
            result.append(int(row[i]))

result2 = []
for row in data2:
    if '율천동' in row[0]:
        for i in range(3, len(row)):
            result2.append(int(row[i]))
    
            
import matplotlib.pyplot as plt
plt.plot(result)
plt.plot(result2)
plt.show()
```
