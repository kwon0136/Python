```python
from bs4 import BeautifulSoup
import pandas as pd

import urllib.request
import urllib.parse

#sel_dates = ['20190118', '20190119', '20190220', '20190221']
#for d in range(len(sel_dates)):
#   print(sel_dates[i])
data = []
for d in pd.date_range('2019-01-18', periods=35):
    #print(d.strftime('%Y%m%d')) # strftime 
    # %Y 2019, %m 01, %d 18
    sel_date = d.strftime('%Y%m%d')
    url = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date=' + sel_date
    
    with urllib.request.urlopen(url) as response:
        html = response.read()
        soup = BeautifulSoup(html, 'html.parser')

    titles = soup.find_all("div",{"class": "tit5"})
    ratings = soup.find_all("td",{"class": "point"})
    #print(titles[0].a.string, ratings[0].string)
    #len(titles)
    for i in range(len(titles)):
        #print(titles[i].a.string, ratings[i].string)
        data.append([titles[i].a.string, ratings[i].string])

data
```

```python
df = pd.DataFrame(data, columns=['title', 'rating'])
#df.shape
#df.columns
df['rating'] = df['rating'].astype(float)
#df.info()
df2 =df.groupby('title')['rating'].mean()
df3 = df2.reset_index()
df4 = df3.sort_values('rating', ascending=False)
df5 = df4.head()
df5
```
